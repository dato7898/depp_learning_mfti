```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
    return 4*x**2 - 3*x +5

def plot_func(f):
    plt.figure(figsize=(10,7))
    x = np.linspace(-1, 2, 100)
    y = list(map(f, x))
    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.scatter([3/8], [F(3/8)], lw=5)
    plt.show()

plot_func(F)
plt.show()
```
